```python
from matplotlib import pyplot as plt

import pandas as pd
import numpy as np

import lizard

plt.style.use('seaborn-colorblind')
plt.ion()
```

```python
# Reading data
df = pd.read_csv('xv6_lizard_analysis.csv')
df.head()
```

```python
# Grouping the function by file and ordering by mean CCN
mean_ccn = df[['FILE', 'CCN']].groupby('FILE').mean().sort_values(by=['CCN'], ascending=False)
mean_ccn.head(5)
```

```python
plt.rcParams['figure.figsize']  = (18, 10)
plt.rcParams['axes.labelsize']  = 20
plt.rcParams['axes.titlesize']  = 20
plt.rcParams['legend.fontsize'] = 20
plt.rcParams['xtick.labelsize'] = 20
plt.rcParams['ytick.labelsize'] = 20
plt.rcParams['lines.linewidth'] = 3

plt.bar(mean_ccn.index,mean_ccn['CCN'])
plt.ylabel('CCN')
plt.xticks(rotation=90)
plt.title('Files x Clomatic Complexity Number')

plt.show()
```

```python
# Most complex functions
filter = df['FILE'] != ".\\usertests.c"
no_tests = df[filter]
functions = no_tests[['FILE', 'FUNCTION_NAME', 'CCN']].sort_values(by=['CCN'], ascending=False)
functions.head(5)
```
